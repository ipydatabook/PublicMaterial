## Boxplots

In the cell below, we make side-by-side boxplots of insurance prices separated by sex.  

```python
ticks = make_array(2.5, 3.5)
labels = make_array("Males", "Females")

plots.boxplot(insurance.where("sex", "male").column("price"), positions = make_array(ticks.item(0)), showmeans=True)
plots.boxplot(insurance.where("sex", "female").column("price"), positions = make_array(ticks.item(1)), showmeans=True)
plots.xticks(ticks, labels)
plots.title("Insurance Prices");
```
